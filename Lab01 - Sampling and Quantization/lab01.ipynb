1)	Utilizando imagem relogio.tif e faça uma reamostragem de seus pixels para 300, 150 e 72 dpi. Sabe-se que a imagem original possui 1250 dpi. Fazer os cálculos por meio do seu algortimo. Não usar o opencv para a conversão

```python
# type your code here
# Python code to read image
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("relogio.tif", cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap='gray', vmin=0, vmax=255)

print('Tamanho da imagem : ',img.shape)
# convertendo para 300dpi

linhas = img.shape[0]
colunas = img.shape[1]

```

```python
p_300 = int(1250/300)
p_150 = int(1250/150)
p_72 = int(1250/72)

print(f" Proporcao para imagem de 300 dpi : {p_300} \n Proporcao para imagem de 150 dpi : {p_150} \n Proporcao para imagem de 72 dpi : {p_72} \n")
```

## Amostragem através da média

```python
#Para 300 dpi usando media

M, N = img.shape
K = p_300
L = p_300

MK = M // K
NL = N // L
new_img = img[:MK*K, :NL*L].reshape(MK, K, NL, L).mean(axis=(1, 3))
plt.imshow(new_img, cmap='gray', vmin=0, vmax=255)
```

```python
#Para 300 dpi usando maximo

M, N = img.shape
K = p_300
L = p_300

MK = M // K
NL = N // L
new_img = img[:MK*K, :NL*L].reshape(MK, K, NL, L).max(axis=(1, 3))
plt.imshow(new_img, cmap='gray', vmin=0, vmax=255)
```

```python
#Para 150 dpi

M, N = img.shape
K = p_150
L = p_150

MK = M // K
NL = N // L
new_img = img[:MK*K, :NL*L].reshape(MK, K, NL, L).mean(axis=(1, 3))
plt.imshow(new_img, cmap='gray', vmin=0, vmax=255)
```

```python
#Para 150 dpi usando maximo

M, N = img.shape
K = p_150
L = p_150

MK = M // K
NL = N // L
new_img = img[:MK*K, :NL*L].reshape(MK, K, NL, L).max(axis=(1, 3))
plt.imshow(new_img, cmap='gray', vmin=0, vmax=255)
```

```python
#Para 72 dpi

M, N = img.shape
K = p_72
L = p_72

MK = M // K
NL = N // L
new_img = img[:MK*K, :NL*L].reshape(MK, K, NL, L).mean(axis=(1, 3))
plt.imshow(new_img, cmap='gray', vmin=0, vmax=255)
```
